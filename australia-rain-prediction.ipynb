Now, we will cap the outliers with upper_fence values.

```python
def cap_outliers(df1, variable, top):
    return np.where(df1[variable]>top, top, df1[variable])

df['rainfall'] = cap_outliers(df, 'rainfall', 3.2)
df['evaporation'] = cap_outliers(df, 'evaporation', 21.8)
df['windgustspeed'] = cap_outliers(df, 'windgustspeed', 99)
df['windspeed9am'] = cap_outliers(df, 'windspeed9am', 55)
df['windspeed3pm'] = cap_outliers(df, 'windspeed3pm', 57)
```

```python
y_train = df_train['raintomorrow']
y_test = df_test['raintomorrow']
```

```python
del df_train['raintomorrow']
del df_test['raintomorrow']
```

Let's prepare the input vectors for model training.

# 7. DictVectorizer

```python
from sklearn.feature_extraction import DictVectorizer
```

```python
train_dicts = df_train.to_dict(orient='records')
```

```python
dv = DictVectorizer(sparse=False)
X_train = dv.fit_transform(train_dicts)
```

```python
dv.get_feature_names_out()
```

These are the list of the feature names of our dataset.

```python
test_dicts = df_test.to_dict(orient='records')
```

```python
X_test = dv.transform(test_dicts)
```

```python
# review the first row of result train vector

row1 = [f"{num:.2f}" for num in X_train[0]]
print(row1)
```

```python
df_train.loc[0,:]
```

```python
# review the first row of result test vector

testrow1 = [f"{num:.2f}" for num in X_test[0]]
print(testrow1)
```

```python
df_test.loc[0,:]
```

# 8. MinMaxScaling

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
# scaling train set and test set

scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# review the first row of test set after scaling

testrow1 = [f"{num:.2f}" for num in X_test[0]]
print(testrow1)
```

```python
# review the first row of train set after scaling

trainrow1 = [f"{num:.2f}" for num in X_train[0]]
print(trainrow1)
```

# 9. Model training

Let's train Logistic Regression with scikit learn

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(random_state=1)
model.fit(X_train, y_train)
```

```python
y_pred_test = model.predict(X_test)
```

# 10. Check accuracy score

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred_test)
```

This is the test set accuracy. Let's compare the train and test set accuracy.

```python
y_pred_train = model.predict(X_train)
```

```python
accuracy_score(y_train, y_pred_train)
```

Both accuracy scores are the same. we can assume that overfitting is not happening here.

# 11. Dummy model

```python
y_pred_test_prob = model.predict_proba(X_test)[:, 1]
```

```python
thresholds = np.linspace(0, 1, 21)
scores = []
for t in thresholds:
    churn_decision = (y_pred_test_prob >= t)
    score = (y_test == churn_decision).mean()
    print('%.2f %.4f' % (t,score))
    scores.append(score)
```

```python
plt.plot(thresholds, scores)
```

Dummy model accuracy is 0.7831. It means when we predict everyday is not raining, we can get accuracy of 0.7831
